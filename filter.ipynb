```python
import csv
months = ['jan.csv','feb_march.csv', 'april.csv', 'may.csv', 'june.csv', 'july.csv',\
          'aug.csv', 'sept.csv', 'oct.csv', 'nov.csv', 'dec.csv']

with open('all_tn_2017_woshape/aug.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    aug = list(reader)
```

```python
for i in range(len(aug[0])):
    if aug[0][i]=='Route':
        route = i
    if aug[0][i]=='BLM':
        blm = i
```

```python
dict={}
for i in months:
    with open('all_tn_2017_woshape/'+i,'r') as csvfile:
        reader = csv.reader(csvfile)
        month = list(reader)
        labels = month[0][:]
        data = month[1:][:]
        for row in data:
            ## Some information is offset, not every row is just 30 rows
            lane_marker = row[blm]
            lane_marker = float(lane_marker)
            start_partition = int(lane_marker)
            end_partition = int(lane_marker + 1)
            location = row[route + len(row) - 30] + ": "+str(start_partition)+"-"+str(end_partition)
            if location in dict:
                dict[location]+=1
            else:
                dict[location]=1
```
