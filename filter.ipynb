```python
for i in range(len(aug[0])):
    if aug[0][i]=='Route':
        route = i
    if aug[0][i]=='Beg LM':
        beg_LM = i
        end_LM = i+1
```

```python
labels = aug[0][:]
data = aug[1:][:]
```

```python
dict={}
```

```python
for i in data:
    ## Some information is offset, not every row is just 30 rows
    location = i[route + len(i) - 30] + ": "+i[beg_LM]+"-"+i[end_LM]
    if location in dict:
        dict[location]+=1
    else:
        dict[location]=1
```
